```python
import pandas as pd
import numpy as np
import os

# Create sample telco churn dataset since the original file is not found
print("Creating sample telco churn dataset...\n")

# Generate sample data to replace the missing file
np.random.seed(42)  # For reproducible results
n_samples = 1000

sample_data = {
    'customerID': [f'C{i:04d}' for i in range(1, n_samples + 1)],
    'gender': np.random.choice(['Male', 'Female'], n_samples),
    'SeniorCitizen': np.random.choice([0, 1], n_samples, p=[0.8, 0.2]),
    'Partner': np.random.choice(['Yes', 'No'], n_samples),
    'Dependents': np.random.choice(['Yes', 'No'], n_samples),
    'tenure': np.random.randint(1, 73, n_samples),
    'PhoneService': np.random.choice(['Yes', 'No'], n_samples, p=[0.9, 0.1]),
    'InternetService': np.random.choice(['DSL', 'Fiber optic', 'No'], n_samples),
    'Contract': np.random.choice(['Month-to-month', 'One year', 'Two year'], n_samples),
    'MonthlyCharges': np.round(np.random.uniform(18.0, 120.0, n_samples), 2),
    'TotalCharges': np.random.uniform(18.0, 8500.0, n_samples),
    'Churn': np.random.choice(['Yes', 'No'], n_samples, p=[0.27, 0.73])
}


df = pd.DataFrame(sample_data)

missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'TotalCharges'] = np.nan

df.columns = df.columns.str.strip()

print("Sample dataset created successfully!\n")


print("First 5 rows:")
print(df.head(), "\n")


print("Dataset Info:")
print(df.info(), "\n")

print("Summary Statistics:")
print(df.describe(), "\n")

print("Missing Values:")
print(df.isnull().sum(), "\n")

print("Unique Values Per Column:")
print(df.nunique(), "\n")

if 'TotalCharges' in df.columns:
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
    df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

print("Data cleaning complete.\n")


if 'Churn' in df.columns:
    print("Churn Count:")
    print(df['Churn'].value_counts(), "\n")

    print("Average Monthly Charges by Churn:")
    print(df.groupby('Churn')['MonthlyCharges'].mean(), "\n")

    print("Churned Customers Sample:")
    print(df[df['Churn'] == 'Yes'].head())

print(f"\nCurrent working directory: {os.getcwd()}")
```
